## ‚úÖ CoT Example ‚Äî Using OpenAI API (`gpt-4` or `gpt-3.5-turbo`)

```python
import openai

openai.api_key = "your-api-key"

prompt = """
Classify the participant's statement into one of the following codes: Supports, Neutral, Opposes (regarding trust in government interventions).

Quote:
"They keep saying it‚Äôs for our safety, but I don‚Äôt feel like anyone‚Äôs actually listening to us."

Let‚Äôs think step by step.
"""

response = openai.ChatCompletion.create(
    model="gpt-4",
    messages=[{"role": "user", "content": prompt}],
    temperature=0
)

print(response['choices'][0]['message']['content'])
```

## ‚úÖ CoT Example ‚Äî Using Hugging Face (`transformers`, e.g. OpenChat)

```python
from transformers import pipeline

cot_pipeline = pipeline("text-generation", model="openchat/openchat-3.5-0106", max_new_tokens=300)

prompt = """
Classify the participant's statement into one of the following codes: Supports, Neutral, Opposes (regarding trust in government interventions).

Quote:
"They keep saying it‚Äôs for our safety, but I don‚Äôt feel like anyone‚Äôs actually listening to us."

Let‚Äôs think step by step.
"""

output = cot_pipeline(prompt)[0]['generated_text']
print(output)
```
